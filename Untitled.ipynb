```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import random
```

```python
def load_data(mode, batch_size):
    x_batch = []
    y_batch = []
        
    for i in range(batch_size):
        if (mode == 'train'):
            file_number = random.randrange(1, 120)
        else:
            file_number = random.randrange(120, 166)
        if (file_number < 10):
            file_name = "00" + str(file_number) + ".jpg"
        elif (file_number < 100):
            file_name = "0" + str(file_number) + ".jpg"
        else:
            file_name = str(file_number) + ".jpg"

        img = cv2.imread('img/{}'.format(file_name))
        mask = cv2.imread('mask/{}'.format(file_name), cv2.IMREAD_GRAYSCALE)
        
        img = cv2.resize(img, (512, 512))
        mask = cv2.resize(mask, (512, 512)) 
            
        x_batch += [img]
        y_batch += [mask]

    x_batch = np.array(x_batch) / 255.
    y_batch = np.array(y_batch)

    yield x_batch, np.expand_dims(y_batch, -1)
```

```python
import keras
from keras.applications.resnet import ResNet50
from keras.applications.vgg16 import VGG16
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D, Dropout, UpSampling2D, Conv2D, MaxPooling2D, Activation


base_model = VGG16(weights='imagenet', input_shape=(512,512,3), include_top=False)


base_out = base_model.output

up = UpSampling2D(32, interpolation='bilinear')(base_out)

conv = Conv2D(1, (1, 1))(up)
conv = Activation('sigmoid')(conv)

model = Model(inputs=base_model.input, outputs=conv)
```

```python
best_w = keras.callbacks.ModelCheckpoint('fcn_best.h5',
                                monitor='val_loss',
                                verbose=0,
                                save_best_only=True,
                                save_weights_only=True,
                                mode='auto',
                                save_freq=1)

last_w = keras.callbacks.ModelCheckpoint('fcn_last.h5',
                                monitor='val_loss',
                                verbose=0,
                                save_best_only=False,
                                save_weights_only=True,
                                mode='auto',
                                save_freq=1)


callbacks = [best_w, last_w]



adam = keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)


model.compile(adam, 'binary_crossentropy')
```

```python
batch_size = 16
model.fit_generator(load_data('train', batch_size),
              steps_per_epoch=100,
              epochs=3,
              verbose=1,
              callbacks=callbacks,
              validation_data=load_data('validation', batch_size),
              validation_steps=50,
              class_weight=None,
              max_queue_size=10,
              workers=1,
              use_multiprocessing=False,
              shuffle=True,
              initial_epoch=0)
```

```python
from albumentations import (
    CLAHE, RandomRotate90, Transpose, RandomCrop, Resize, ShiftScaleRotate, Blur, OpticalDistortion, 
    GridDistortion, HueSaturationValue, IAAAdditiveGaussianNoise, GaussNoise, MotionBlur, 
    MedianBlur, IAAPiecewiseAffine, IAASharpen, IAAEmboss, RandomContrast, RandomBrightness, 
    Flip, HorizontalFlip, OneOf, Compose, PadIfNeeded, LongestMaxSize, PadIfNeeded, ElasticTransform,Cutout
)
```

```python
def strong_aug(p=1.0):
    return Compose([
        ShiftScaleRotate(shift_limit=0.125, scale_limit=0.2, rotate_limit=10, p=0.7, border_mode=cv2.BORDER_CONSTANT),
        #RandomCrop(512, 512),
        #PadIfNeeded(min_height=224, min_width=224, border_mode=cv2.BORDER_CONSTANT, p=1.0),
        #Resize(64, 64),
        #RandomRotate90(),
        ElasticTransform(1.), 
        HorizontalFlip(),
        #Cutout(p=1.),
        #Transpose(),
        OneOf([
            OpticalDistortion(p=0.3),
            GridDistortion(p=0.1),
        ], p=0.5),
        OneOf([
            CLAHE(clip_limit=3)
        ], p=0.4),
        HueSaturationValue(p=0.7),
         
        
        
    ],
        p=p)
```

```python
augmentation = strong_aug(p=1.0)

```

```python
for i in range(1, 84):
    if (i < 10):
        file_name = "0" + str(i) + ".jpg"
    else:
        file_name = str(i) + ".jpg"
    img = cv2.imread('img/{}'.format(file_name))
    mask = cv2.imread('mask/{}'.format(file_name), cv2.IMREAD_GRAYSCALE)
    
    #data = {'image': np.array(img), 'mask': np.array(mask)}
    augmented = augmentation(image=img, mask=mask)
    crop_img, crop_mask = augmented["image"], augmented["mask"]
    
    cv2.imwrite('img/aug' + str(i) + '.jpg', crop_img)
    cv2.imwrite('mask/aug' + str(i) + '.jpg', crop_mask)
```
